```python
datos.workclass.value_counts()
```

```python
## Reemplacemos los datos nulos 

## Cambiamos los nulos de workclass por su moda

wc_nulos_indices = datos.workclass[datos.workclass.isnull()].index ## obtenemos los índices de los nulos

##reemplazamos los datos nulos por la moda
datos.loc[wc_nulos_indices , "workclass" ] = datos.workclass.mode().values[0]  
```

```python
datos.workclass.mode().values[0]
```

```python
datos.workclass.value_counts()
```

```python
## Cambiamos los nulos de occupation por su moda

datos.occupation.value_counts()
```

```python
occu_nulos_indices = datos.occupation[datos.occupation.isnull()].index
datos.loc[occu_nulos_indices, "occupation"] = datos.occupation.mode().values[0]
```

```python
## Para eliminar los datos nulos de native_country utilizaremos tres métodos diferentes

datos2 = datos
datos3 = datos
```

```python
## Para eliminar los nulos de native_country elinimaremos las filas

datos = datos.dropna()     ## elimina las filas con nulos, deja los índices iguales
```

```python
## Para reorganizar los índices

datos.reset_index(drop = True , inplace = True)
```

```python
datos.tail()
```

```python
## Llenaremos los datos nulos con los valores posteriores

datos2 = datos2.fillna(method = "ffill")

datos2.isnull().sum()
```

```python
## Llenaremos los datos nulos con los valores anteriores

datos3 = datos3.fillna(method = "bfill")

datos3.isnull().sum()
```
