```python
## Primero hacemos la importación de la librería

import pandas as pd 
```

```python
## Creamos una Serie 

ciudades = ["Santa marta", "medellín", "BoGotá", "CaLi", "cartagEna", "BucarAManga"]

serie = pd.Series(ciudades)
#print (serie)

## Extraemos información de la Serie

if (serie[0]== "Cali"):
    print ("Hola mundo")
else:
    print ("La ciudad es " +serie[0])
    
## Agregamos información a la Serie

serie[6] = "Pasto"
#print (serie)

## Cambiamos los índices de la Serie

serie.index = ["A", "B", "C", "D", "E", "F", "G"]
print (serie)

## Agregando y cambiando elementos

serie["H"] = "MiTú"
serie["A"] = "Barranquilla"

## Manipulación de los datos

serie = serie.str.capitalize()

serie.loc["D":"F"]  ## Muestra los datos desde el índice D hasta el F

len(serie)
serie
```

```python
## Creamos un DataFrame

numeros = {"pares":list(range(0,200,2)),"impares": list(range(1, 200, 2)) }

datos = pd.DataFrame(numeros)

## Extraemos información del DataFrame

print (datos.info())       ## Muestra la información general del DataFrame

print (datos.describe())   ## Muestra la información general matemática de los datos numéricos

print (datos.shape)        ## Muestra la información de las columnas y las filas del DataFrame

print (datos.head(8))      ## Muestra los primeros 8 datos del DataFrame, por defecto muestra 5

print (datos.tail(10))     ## Muestra los últimos 10 datos del DataFrame, por defecto muestra 5

print (datos.to_string())  ## Muestra todo el DataFrame   

print (datos.loc[11:27])   ## Muestra los datos desde el índice 11 hasta el 27
```

```python
## Exploración básica de un DataFrame

## Leemos el archivo de Internet
camino = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
datos = pd.read_csv(camino, header=None) ## creamos un DataFrame con los datos y le indicamos no leer los headers

# Creación de un header

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

datos.columns = headers  ## asignamos el header de acuerdo a los nombre que creamos anteriormente
datos.head(10)           ## vemos los primeros datos del DataFrame
```
