```python
## Importamos las librerías

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Leemos el archivo

datos = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv')
```

```python
## Veamos el comportamiento del precio

datos.price.hist()      ## Histograma del precio
```

```python
datos.price.plot(kind = 'box')   ## Diagrama de cajas
```

```python
## Elijamos una variable categórica y veamos cómo funciona el método value_counts()

datos['drive-wheels'].value_counts()

## Tenemos 118 de tracción delantera, 75 de tracción tracera y 8 de tracción de cuatro ruedas
```

```python
## Hagamos una gráfica circular de las tracciones

datos['drive-wheels'].value_counts().plot(kind = 'pie', figsize = (10,10), autopct='%1.1f%%')
```

```python
## Veamos cómo funciona el método value_counts() para variables numéricas

datos.width.value_counts()
```

```python
## Ahora, haremos uso de la librería seaborn para graficar

## Histograma del precio

fig, ax = plt.subplots(figsize = (15,8)) ## Definimos la figura

sns.histplot(data = datos, x = 'price' , hue = 'drive-wheels' , ax = ax)
```

```python
## Diagrama de cajas

fig, ax = plt.subplots(figsize = (15, 8))
sns.boxplot(data = datos, y = 'price' , x = 'drive-wheels' , ax = ax)
```

```python
## Podemos ver además de graficamente la comparación de las medias del precio respecto al tipo de tracción

datos.groupby('drive-wheels').mean()['price']
```

```python
datos.groupby('drive-wheels').median()['price']  ## Vemos el comportamiento de las medianas
```

```python
## Veamos la apreciencia de los cuartiles

q1 = datos.groupby('drive-wheels').quantile(0.25)['price']
q1  ## primer cuartil
```

```python
q2 = datos.groupby('drive-wheels').quantile(0.5)['price']
q2
```

```python
q3 = datos.groupby('drive-wheels').quantile(0.75)['price']
q3
```

```python
q4 = datos.groupby('drive-wheels').quantile(1)['price']
q4
```

```python
datos.groupby('drive-wheels').describe()['price']
```

```python
## Y para el rango intercuartil tenemos que

iqr = q3 - q1
iqr
```

```python
datos.groupby('drive-wheels').max()['price']
```

```python
## Veamos qué ocurre con los extremos
## Extremo superior

UP = q3 + 1.5*iqr
U_prueba_superior = datos.groupby('drive-wheels').max()['price']

UP < U_prueba_superior
```

```python
UP
```

```python
## Extremo inferior

LP = q1 - 1.5*iqr
U_prueba_inferior = datos.groupby('drive-wheels').min()['price']

LP > U_prueba_inferior
```

```python
U_prueba_inferior
```

```python
## Hagamos diagramas de cajas con otros variables: body-style y engine-location

figure, ax = plt.subplots(1,2, figsize = (13,8))
sns.boxplot(y = 'price' , x = 'body-style' , data = datos , ax = ax[1])
sns.boxplot(y = 'price' , x = 'engine-location', data = datos , ax = ax[0])
```

```python
## Hagamos diagramas de cajas con otras variables: make 

figure, ax = plt.subplots(figsize = (17,8))

sns.boxplot(y = 'price' , x = 'make' , data = datos , ax = ax)
```
