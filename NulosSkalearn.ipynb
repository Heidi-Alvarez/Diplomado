```python
## Importamos las librerias

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
```

```python
## Importamos los datos

datos = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv", names = headers) 
```

```python
## Reemplacemos los valores ? en normalized-losses por NaN

datos = datos.replace('?', np.nan)
```

```python
## Ahora veamos cuántos datos nos faltan

faltantes = datos.isnull().sum()

faltantes[faltantes != 0].sort_values(ascending = False)

```

```python
## Reemplazaremos por la moda en: normalized-losses, stroke, bore, horsepower, peak-rpm
## Primero calculamos el valor promedio de cada uno de ellos

nor_lss_promedio = datos["normalized-losses"].astype(float).mean(axis=0)
ske_promedio = datos["stroke"].astype(float).mean(axis=0)
bre_promedio = datos["bore"].astype(float).mean(axis=0)
hs_pw_promedio = datos["horsepower"].astype(float).mean(axis=0)
pk_rpm_promedio = datos["peak-rpm"].astype(float).mean(axis=0)

## Luego reemplazamos cada uno de los valores en el DataFrame

datos["normalized-losses"].replace(np.nan, nor_lss_promedio, inplace = True)
datos["stroke"].replace(np.nan, ske_promedio , inplace = True)
datos["bore"].replace(np.nan, bre_promedio, inplace = True)
datos["horsepower"].replace(np.nan, hs_pw_promedio , inplace = True)
datos["peak-rpm"].replace(np.nan, pk_rpm_promedio , inplace = True)
```

```python
## Reemplazaremos por la moda num-of-doors
## Primero calculamos la moda

num_ds_mode = datos["num-of-doors"].mode()

num_ds_mode
```

```python
## Reemplazamos en el DataFrame

datos["num-of-doors"].replace(np.nan, "four" , inplace = True)
```

```python
## Para price eliminaremos las filas sin datos. La razón es que eventualmente buscaremos predecir este dato

datos.dropna(subset = ["price"] , axis = 0 , inplace = True)
```

```python
## Verificando la información del DataFrame

datos.info()
```

```python
## Dado que no todas las columnas tienen el tipo correcto, haremos las correcciones pertinentes

datos[["bore", "stroke", "price", "peak-rpm"]] = datos[["bore", "stroke", "price", "peak-rpm"]].astype(float)
datos[["normalized-losses"]] = datos[["normalized-losses"]].astype(int)
```

```python
datos.dtypes ## Vemos nuevamente el tipo de dato y verificamos que sean los correctos
```

```python
## Ahora, veamos cómo son los datos de la columna make

datos.make.unique()
```

```python
## Estandaricemos los datos
## Crearemos una nueva columna que nos dé millas por galón mpg a litros por 100 km
## La conversión es L/100km = 235/mpg

datos["city-L/100km"] = 235/datos["city-mpg"]  ## Agrega una columna con la conversión
datos["highway-L/100km"] = 235/datos["highway-mpg"] 
datos.head()
```

```python
## Otra estandarización es dejar los valores en un rango entre 0 y 1

datos["length"] = datos["length"]/datos["length"].max()
datos["width"] = datos["width"]/datos["width"].max()
datos["height"] = datos["height"]/datos["height"].max()
```
